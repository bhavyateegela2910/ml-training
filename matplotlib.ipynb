```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,9,3,8,4])
y=np.array([2,5,10,8,4])
plt.plot(x,y,marker='o',ms=15,mec='green',mfc="green",linestyle='-.',color='blue',lw=9)
plt.xlabel("score",loc='left')
plt.ylabel("subjects",loc="top")
plt.title("students score",loc="right",fontsize=18,color='red')
plt.grid(color='pink',linewidth=2,linestyle='--')
plt.show()
```

```python
x=np.array([1,7,4,5,7,8,9,10,9,5])
y=np.array([2,6,7,2,5,7,3,10,11,12])
c=np.array(['red','pink','b','k','purple','orange','g','brown','yellow','gray'])
sizes=np.array([60,90,120,100,60,30,50,110,30,90])
plt.scatter(x,y,marker='P',color=c,s=sizes)
plt.xlabel("score",loc='left')
plt.ylabel("subjects",loc="top")
plt.title("students score",loc="right",fontsize=18,color='red')
plt.grid(color='pink',linewidth=2,linestyle='--')
plt.show()
```

```python
x=np.array(["python","c","c++","java"])
y=np.array([3,1,8,4])
plt.bar(x,y)
plt.show()
```

```python
y=np.array([35,25,25,25])
mylabel=['apple','mango','banana','dates']
myexplode=[0,0.5,0,0]
color=['orange','pink','black','yellow']
plt.pie(y,labels=mylabel,explode=myexplode,shadow=True,colors=color)
plt.legend(title="fruits",loc='upper left')
plt.show()
```
